```python
import pandas as pd
import os
from pathlib import Path

DATA_DIR = Path("../data")

city_files = [f for f in DATA_DIR.glob("*.csv")]

all_cities = []

for file in city_files:
    df = pd.read_csv(file)
    df["city"] = file.stem  # add city name
    all_cities.append(df)

data = pd.concat(all_cities, ignore_index=True)
print("Total rows loaded:", len(data))

# ---- Basic cleaning ----
# remove rows with missing price or bedrooms
data = data.dropna(subset=["price", "bedrooms"])

# convert price to numeric if needed
data["price"] = data["price"].astype(float)

# save cleaned file
data.to_csv("../data/combined_clean.csv", index=False)
print("Saved cleaned dataset â†’ data/combined_clean.csv")

data.head()
```
